# Example 3.3

Find the maximum dry density of a soil compacted with 6% of water content and $G_s=2.66$. Plot the ZAV and lines with saturation of 80% and 60%.

## Data:

$w=6\%$

$G_s=2.66$

## Principles:

$\gamma_d=\cfrac{\gamma_w G_s}{1+wG_s/S}$

$\gamma_{d,max}=\cfrac{\gamma_w G_s}{1+wG_s}$

```python
import numpy as np
import matplotlib.pyplot as plt
import Soilved as slv
```

```python
# I created a function for the first equation in principles
w=6 # water content in percent
S=100 # saturation in percent

gamma_dmax=slv.GetCompactionLines(w, S, Gs=2.66) # returns max dry unit weight
print("gamma_dmax= %1.2f kN/m^3" %gamma_dmax)
gamma_dmax=slv.GetCompactionLines(w, S, gamma_w=62.4,Gs=2.66) # returns max dry unit weight
print("gamma_dmax= %1.2f pcf" %gamma_dmax)
```

```python
# Plot:
# I first create an array with several water contents

w=np.linspace(5,25,100) #100 w values from 5 to 25 percent

#Now I evaluate the ZAV (i.e., S=100%)

ZAV=slv.GetCompactionLines(w, S=100, Gs=2.66) #gets the maximum possible dry densities

#now I can plot the results
plt.plot(w, ZAV)
plt.xlabel(r'Water content $w$ [%]')
plt.ylabel(r'Max. dry unit weight $\gamma_{d, max}$ [kN/m$^3$]')
```

```python
# Now I can repeat for S=80% and S=60% and plot in a single graph

l_60=slv.GetCompactionLines(w, S=60, Gs=2.66) #gets line of equal saturation 80%

l_80=slv.GetCompactionLines(w, S=80, Gs=2.66) #gets line of equal saturation 80%

#plot
plt.plot(w, ZAV) #plots ZAV
plt.plot(w, l_80) #plots line of 80%
plt.plot(w, l_60) #plots line of 60%
plt.xlabel(r'Water content $w$ [%]')
plt.ylabel(r'Max. dry unit weight $\gamma_{d, max}$ [kN/m$^3$]')
```
