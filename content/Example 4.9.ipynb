# Example 4.9

The figure shows water flow through the soil specimen in a cylinder. The specimen's $k$ is $3.4 \times 10^{-4}$ cm/s. 
1. Calculate total head, pressure head, and pore pressure at point A and B.
2. Draw the levels of water in standpipes.
3. Compute the amount of water flow $q$ through the specimen.



## Data:

Elevations and figure

## Principles:

$h=z+u/\gamma_w$

$h_w=u/\gamma_w$

$q=kiA$

$i=\Delta h/L$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# We start defining the boundary conditions at points A and B
h_A=0.28 # head at point A in m
h_D=0.2 #Head at point B in m

# Now we repeat for point D
# total head, pressure head, and pore pressure at point A
h_w=0.28-0.05 #pressure head at A
print('hw= %1.2f m' % h_w)

print('hA= %1.2f m' % h_A)

u=h_w*9.81 #pore pressure at A
print('u_A= %1.2f kPa' %u)
```

```python
# total head, pressure head, and pore pressure at point A
h_w=0.20-0.125 #pressure head at D
print('hw= %1.2f m' % h_w)

print('hD= %1.2f m' % h_D)

u=h_w*9.81 #pore pressure at D
print('u_A= %1.2f kPa' %u)
```

```python
dh=h_A-h_D #change of head in m

L=3*60/1000 # total Length of soil
i=dh/L

print('i= %1.2f' %i)
```

```python
# For the interior points use i to calculate drop of h
L_AB=0.06 #length between A and B
dh_AB=i*L_AB #head drop
h_B=h_A-dh_AB #head at B

z_B=0.075 # elevation of point B in m
h_w=h_B-z_B #pressure head at point B
u=h_w*9.81 #pore pressure @ B

print('hw= %1.2f m' % h_w)
print('hB= %1.2f m' % h_B)
print('u_B= %1.2f kPa' %u)
```

```python
# We can repeat the process for point C
L_BC=0.06 #length between B and C
dh_BC=i*L_BC #head drop
h_C=h_B-dh_BC #head at C

z_C=0.1 # elevation of point C in m
h_w=h_C-z_C #pressure head at point C
u=h_w*9.81 #pore pressure @ B

print('hw= %1.2f m' % h_w)
print('hC= %1.2f m' % h_C)
print('u_C= %1.2f kPa' %u)
```

```python
# Draw levels of standpipes
h=[h_A, h_B, h_C, h_D] #list of heads
plt.plot(h, '*')
```

```python
# now let's calculate the cross-sectional area

A=0.25*np.pi*(60/1000)**2 #A=pi*D^2/4

print('A= %1.2e m^2' %A)
```

```python
# Finally use Darcy's law to calculate the flow
k=3.4e-6 #permeability in m/s
q=k*i*A
print('q= %1.2e m^3/s' % q)
print('q= %1.2e lt/day' % (q*3680000*24))
```
