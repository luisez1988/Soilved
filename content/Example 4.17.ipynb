# Example 4.17

A permeable water channel will be constructed as shown in the figure below. Three layers of sand with permeabilities $k_1=1.5 \times 10^{-5}$ cm/s, $k_2=7.89 \times 10^{-3}$ cm/s, and $k_3=1.2 \times 10^{-4}$ cm/s connects a river and the proposed channel as illustrated in the figure. Calculate the hydraulic gradient between the river and the channel, the equivalent permeability of the layers, and the flow per unit of length in the channel.



## Data:

$k_1= 1.5 \times 10^{-5}$ cm/s

$k_2= 7.89 \times 10^{-3}$ cm/s

$k_3= 1.2 \times 10^{-4}$ cm/s

$t_1=3$ ft

$t_2=3$ ft

$t_3=4$ ft

$W=1$ ft (dimension going out of the screen)

Figure

## Principles

$k_{eq}=\cfrac{\sum_{i=1}^n k_iA_i}{A}$

$i= \Delta h/L$

$h=z+u/\gamma_w$

$q=kiA$

```python
import numpy as np
```

```python
# input
k1=1.5e-5/(2.54*12) # k1 in ft/s
k2=7.89e-3/(2.54*12) #k2 in ft/s
k3=1.2e-4/(2.54*12) # K3 in ft/s

A1=3*1 #cross sectional area soil 1 ft2/ft
A2=3*1 # cross sectional area soil 2 ft2/ft
A3=4*1 # cross sectional area soil 3 ft2/ft

At=10*1 # total cross sectional area ft2/ft
```

```python
# now I can calculate k_eq
k_eq=(k1*A1+k2*A2+k3*A3)/At # equivalent permeability in ft/s
print('k_eq= %1.2e ft/s' %k_eq)
```

```python
# We need the head at the entrance and exit of the soil
h_enter=137.23 #water elevation at entrance in ft
h_exit=125.5 #water elevation at exit in ft

dh=h_enter-h_exit #difference of head
print('dh= %1.2f ft' % dh)
```

```python
# we can calculate the slope angle and then the distance
beta=np.arctan(dh/1255.6) #slope angle in rad.
print('beta= %1.2f deg' % (beta*180/np.pi))

L=1255.6/np.cos(beta) # inclined length in ft
print('L= %1.2f ft' %L)
```

```python
# Now we can calculate the hydraulic gradient
i=dh/L #hydraulic gradient
print('i= %1.2e' %i)

# And finally the flow
q=k_eq*i*At
print('q= %1.2e ft^3/s/ft' %q)
```
