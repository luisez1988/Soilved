# Example 4.8

Calculate the hydraulic gradient between points A and B in the figure bellow and plot the piezometric line between points C and D.



## Data:

Elevations in Figure

## Principles:

$h=z+u/\gamma_w$

$i=\Delta h/ L$

```python
# For points A we can start defining z
z_A=65 #elevation of point A in m
hw_A=100-65 #water head over point A
h_A=z_A+hw_A #Total head @ A

#For point B it is similar
z_B=56 #elevation of point B in m
hw_B=200-56 #water head over point A
h_B=z_B+hw_B #Total head @ A

#Now we can calculate the difference of head Delta h

Dh=h_B-h_A # difference of head
print('Dh= %1.1f m' %Dh)
```

```python
# The drop of pressure is produced between A and B thus:
L=z_A-z_B # Length in m

# finally we can calculate the hydraulic gradient

i=Dh/L #hydraulic gradient (dimensionless)
print('i=%1.2f' %i)
```
